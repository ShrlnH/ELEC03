```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
crimes = pd.read_csv('datasets\\Chicago_Crimes.csv')
```

```python
crimes
```

```python
crimes.describe()
```

<h2>Checking Null Values</h2>

```python
crimes.isnull()
```

```python
crimes.isnull().sum()
```

```python
crimes['Date'] = crimes['Date'].astype(str)
crimes['Date'] = crimes['Date'].replace('/', '-') 
crimes['Date'] = crimes['Date'].str.strip()
crimes['Date'] = pd.to_datetime(crimes['Date'], dayfirst=True, errors='coerce')

crimes['Location Description'] = crimes['Location Description'].fillna('unknown_location')
crimes['Community Area'] = crimes['Community Area'].fillna('unaccounted')
crimes['X Coordinate'] = crimes['X Coordinate'].fillna('unaccounted')
crimes['Y Coordinate'] = crimes['Y Coordinate'].fillna('unaccounted')
crimes['Latitude'] = crimes['Latitude'].fillna(crimes['Latitude'].mode()[0])
crimes['Longitude'] = crimes['Longitude'].fillna(crimes['Longitude'].mode()[0])
crimes['Location'] = crimes['Location'].fillna('unaccounted')

crimes['Year'] = crimes['Date'].dt.year
crimes['Month'] =  crimes['Date'].dt.month
crimes['Day'] =  crimes['Date'].dt.day
crimes['DayOfWeek'] =  crimes['Date'].dt.dayofweek
crimes['Hour'] =  crimes['Date'].dt.hour
```

```python
crimes.isnull().sum()
```

<h2>12.Which month on the year 2025 has the most robbery incident recorded?</h2>

```python
rob_2025 = crimes[(crimes['Primary Type'] == 'ROBBERY') & (crimes['Year'] == 2025)]
count_2025 = rob_2025.groupby('Month').size()
most_robbery_month = count_2025.idxmax()
most_robbery_count = count_2025.max()
count_2025
```

```python
plt.figure(figsize=(10, 5))
count_2025.plot(marker='o', title='Monthly Robbery Incidents in Chicago (2025)')

plt.scatter(most_robbery_month, most_robbery_count, color='red', zorder=5)
plt.text(most_robbery_month, most_robbery_count + 5, 'Highest', color='red')
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

<h2>INSIGHTS:</h2>

<h3>32. January has the highest number of incidents at 374, which is noticeably larger than any other month, indicating that criminal activity peaks at the start of the year.</h3>

<h3>33.There is a significant decrease in incidents from January (374) to February (294), showing a drop of about 21%, which suggests a strong decline in criminal activity after the start of the year.</h3>

<h3>34. Incident numbers reach their lowest point in July with only 48 cases, suggesting mid-year tends to have the least criminal activity.</h3>

<h3>35.From April to December, incident numbers remain consistently low, ranging between 44 and 68, which suggests a period of relative stability in criminal activity for most of the year.</h3>
